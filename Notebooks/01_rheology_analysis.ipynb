```python
import pandas as pd 
```

```python


# Load using correct encoding
df = pd.read_csv(r"../data/Book.csv",encoding="latin1")

# Show first 10 rows
df.head(10)
```

```python
df = df.drop([0, 1])  # delete row 0 and 1
df = df.reset_index(drop=True)
df.head()
```

```python
cols = ["Point No.", "Shear Rate", "Shear Stress", "Viscosity", "Temperature", "Torque"]

for c in cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")
```

```python
df
```

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

```python
df = df.dropna(how="all")
```

```python
df = df.dropna(subset=["Shear Rate", "Shear Stress", "Viscosity"])
df = df.reset_index(drop=True)
```

```python
df.head(20)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.plot(df["Shear Rate"], df["Shear Stress"], marker='o')
plt.xlabel("Shear Rate (1/s)")
plt.ylabel("Shear Stress (Pa)")
plt.title("Flow Curve: Shear Stress vs Shear Rate")
plt.grid(True)
plt.show()
```
