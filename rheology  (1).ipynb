```python
import numpy as np

# Extract data
x = df["Shear Rate"].astype(float)
y = df["Shear Stress"].astype(float)

# Remove any zero or negative values (log cannot process)
mask = (x > 0) & (y > 0)
x = x[mask]
y = y[mask]

# Log-transform
logx = np.log(x)
logy = np.log(y)

# Linear regression on log-log data
coeff = np.polyfit(logx, logy, 1)
n = coeff[0]             # Flow behavior index
logK = coeff[1]
K = np.exp(logK)         # Consistency index

print("Flow behavior index n =", n)
print("Consistency index K =", K)
```

```python
# Predicted stress using fitted model
y_pred = K * x**n

plt.figure(figsize=(6,4))
plt.loglog(x, y, 'o', label="Experimental Data")
plt.loglog(x, y_pred, '-', label=f"Power-law Fit (K={K:.3e}, n={n:.3f})")

plt.xlabel("Shear Rate (1/s)")
plt.ylabel("Shear Stress (Pa)")
plt.title("Power-law Fit on Log-Log Scale")
plt.grid(True, which="both")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def HB_model(gamma, tau0, K, n):
    return tau0 + K * (gamma ** n)
```

```python
gamma = df["Shear Rate"].astype(float).values       # shear rate
tau = df["Shear Stress"].astype(float).values       # shear stress

# Keep only positive values (HB cannot fit zeros)
mask = (gamma > 0) & (tau > 0)
gamma = gamma[mask]
tau = tau[mask]
```

```python
# initial guess: tau0, K, n
initial_guess = [1, 100, 0.1]

popt, pcov = curve_fit(HB_model, gamma, tau, p0=initial_guess, maxfev=50000)

tau0, K_HB, n_HB = popt

print("Yield Stress (tau0) =", tau0)
print("Consistency Index (K) =", K_HB)
print("Flow Index (n) =", n_HB)
```

```python
gamma_fit = np.linspace(min(gamma), max(gamma), 200)
tau_fit = HB_model(gamma_fit, tau0, K_HB, n_HB)

plt.figure(figsize=(6,4))
plt.loglog(gamma, tau, 'o', label="Experimental Data")
plt.loglog(gamma_fit, tau_fit, '-', label="HB Fit")

plt.xlabel("Shear Rate (1/s)")
plt.ylabel("Shear Stress (Pa)")
plt.title("Herschel–Bulkley Fit")
plt.grid(True, which="both")
plt.legend()
plt.show()
```
